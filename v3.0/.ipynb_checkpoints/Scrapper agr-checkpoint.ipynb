```python
import numpy as np
import pandas as pd

import requests, bs4
from datetime import datetime
import math

from IPython.display import clear_output
```

```python
dataset = pd.read_csv('data.csv')
```

```python
dataset.head()
```

```python
len(dataset)
```

```python
def alexaGlobalRank(x):
    try:
        res = requests.get('https://www.alexa.com/siteinfo/'+x)
        if res.status_code == requests.codes.ok:
            ressoup = bs4.BeautifulSoup(res.text)
            return int(ressoup.select('.rankmini-rank')[0].getText().strip().replace('#', '').replace(',', ''))**-1
        else:
            return -1
    except Exception as e:
        return e
```

```python
agr = []
for i, x in enumerate(dataset['URL']):
    agr.append(alexaGlobalRank(x))
    print(i, x, agr[-1])
```

```python
print(len(agr))
print(agr)
```

```python
dataset['alexaGlobalRank'] = agr
```
