```python
import os
import pickle
```

```python
domains = []
```

```python
for folderName, subfolders, filenames in os.walk('blacklists'):
    print('The current folder is ' + folderName)

    for subfolder in subfolders:
        print('SUBFOLDER OF ' + folderName + ': ' + subfolder)
    for filename in filenames:
        print('FILE INSIDE ' + folderName + ': '+ filename)
        f = open(folderName+'\\'+filename)
        content = f.read()
        if filename == 'domains':
            domains += content.split('\n')
    print('')
```

```python
domains = list(set(domains))
```

```python
pickle.dump(domains, open('BlackList.pickle', 'wb'))
```
