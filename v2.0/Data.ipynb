# Libraries

```python
import numpy as np
import pandas as pd
from sklearn.utils import shuffle
```

# Data

```python
clean_url = pd.read_csv('../DataSet/CleanURL.csv')
```

```python
clean_url.head()
```

```python
clean_url['URL'] = clean_url['URL'].apply(lambda x: x if x.startswith('https://') or x.startswith('http://') else 'https://'+x)
```

```python
malicious_url = pd.read_csv('../DataSet/MaliciousURL.csv')
```

```python
malicious_url.head()
```

```python
clean_url['Label'] = 0
```

```python
malicious_url['Label'] = 1
```

```python
f= open('../DataSet/URL.txt', 'r')
```

```python
cont = f.read()
```

```python
url = []
for c in cont.split('\n'):
    url.append(c.split(','))
```

```python
url
```

```python
url = pd.DataFrame(url, columns=('URL', 'Label'))
```

```python
url.head()
```

```python
data = pd.concat([clean_url, malicious_url, url])
```

```python
data = shuffle(data.drop_duplicates()).reset_index(drop=True)
```

```python
data.to_csv('data.csv', index=False)
```
