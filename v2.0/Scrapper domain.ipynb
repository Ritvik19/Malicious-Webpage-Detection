```python
import numpy as np
import pandas as pd
```

```python
import requests, bs4
import pickle
```

```python
d = pickle.load(open('../BlackList.pickle', 'rb'))
```

```python
dataset = pd.read_csv('data.csv')
```

```python
def domainBlackListed(x):
    try:
        res = requests.get('https://www.whois.com/whois/'+x)
        if res.status_code == requests.codes.ok:
            websitesoup = bs4.BeautifulSoup(res.text)
            if websitesoup.select('.df-value')[0].getText() in d:
                return 1
            else:
                return 0
        else:
            return -1
    except:
        return -1
```

```python
dbl = []
for i, x in enumerate(dataset['URL']):
    dbl.append(domainBlackListed(x))
    print(i, dbl[-1])
```
